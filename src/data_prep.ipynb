```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv("diabetes_data_upload.csv")
```

```python
df = df.rename(columns = {'class': 'diabetes_class'})

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_') # standardising column names

binary_map = {'Yes': 1, 'No': 0}

binary_cols = [col for col in df.columns if df[col].dropna().isin(['Yes', 'No']).all()]
df[binary_cols] = df[binary_cols].replace(binary_map) # changing binary columns from yes/no to 1/0

df['gender'] = df['gender'].replace({'Male': 1, 'Female': 0}) # changing binary column for gender
df['diabetes_class'] = df['diabetes_class'].replace({'Positive': 1, 'Negative': 0}) # changing binary column for target variable 

scaler = StandardScaler() # scaling age variable
df['age_scaled'] = scaler.fit_transform(df[['age']]) 
df = df.drop(columns=['age'])

df.to_csv('diabetes_data_clean.csv', index = False)
```

```python
print("Class Distribution:\n", df['diabetes_class'].value_counts(normalize=True)) # checking class imbalance

# testing for multicollinearity using VIF
mc = df.drop(columns=['diabetes_class', 'age_scaled']) # selecting only independent features
mc_vif = mc.assign(const=1) # adding constant / intercept column

vif_df = pd.DataFrame()
vif_df["feature"] = mc_vif.columns
vif_df["VIF"] = [variance_inflation_factor(mc_vif.values, i) for i in range(len(mc_vif.columns))]

print("\nVIF Results (Sorted):")
print(vif_df.sort_values(by="VIF", ascending=False))
```
